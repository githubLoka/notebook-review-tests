```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(0)

# Create a DataFrame with randomly generated data
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
     'data1': range(6),
     'data2': np.random.randint(0, 20, 6)},
    columns = ['key', 'data1', 'data2'])

# Print the DataFrame
df
```

## this is markdown text

```python
df.plot(kind='scatter', x='data1', y='data2')
```

```python
df.plot(kind='box')
```

```python
df.plot(kind='line')
```
